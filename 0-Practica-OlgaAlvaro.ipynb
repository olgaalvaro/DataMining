Vamos a agrupar los barrios por metros cuadrados. Podemos usar una matriz de similaridad de Tukey tal y como hicimos en el curso de estadística:

```r
tky<-TukeyHSD(aov( formula=Square.Meters~Neighbourhood, data=df_madrid ))
tky.result<-data.frame(tky$Neighbourhood)
cn <-sort(unique(df_madrid$Neighbourhood))
resm <- matrix(NA, length(cn),length(cn))
rownames(resm) <- cn
colnames(resm) <- cn
resm[lower.tri(resm) ] <- round(tky.result$p.adj,4)
resm[upper.tri(resm) ] <- t(resm)[upper.tri(resm)] 
diag(resm) <- 1
library(ggplot2)
library(reshape2)
dfResm <- melt(resm)

options(repr.plot.height=6,repr.plot.width=6,repr.plot.res = 300)
ggplot(dfResm, aes(x=Var1, y=Var2, fill=value))+
  geom_tile(colour = "black")+ 
  scale_fill_gradient(low = "white",high = "steelblue")+
  ylab("Class")+xlab("Class")+theme_bw()+
  theme(axis.text.x = element_text(angle = 90, hjust = 1),legend.position="none")
```

Vamos a crear una nueva columna en el dataframe df_madrid con un nuevo identificador marcado por los clusters obtenidos. Esta columna la llamaremos neighb_id

```r
# También podemos aplicar cutree_1h ya que conserva el orden del dendrograma que hemos dibujado antes:
cls<-cutree_1h.dendrogram(hcd,h=0.9)

cls<-factor(cls)
cls
```

Vamos a crear dos grupos, uno test y otro train.

Tratamos de predecir los metros cuadrados en función del resto de columnas del dataframe.

```r
# Crear dos grupos, uno para test y otro para train

set.seed(1234)    # semilla
idx<-sample(1:nrow(df_madrid),nrow(df_madrid)*0.7)  # muestra aleatoria 

# Dividimos la muestra del data frame df_madrid en train (entrenamiento) y test
df_madrid.train<-df_madrid[idx,]
df_madrid.test <-df_madrid[-idx,]  # Negativo porque son los datos no training

paste ("Nº de muestras para train: ", nrow(df_madrid.train))  # 70% training
paste ("Nº de muestras para test: ", nrow(df_madrid.test) )   # 30% testing
```

```r
# Modelo lineal (tras probar con muchos este modelo parece que es el q mayor R2 y menor AIC)
# Las variables utilizadas tienen valores significativos
# R² cuanto más cercano a  1  mejor seguirá la predicción a los datos reales.

# Square.Metere = 𝛽0 + 𝛽1 * Bedrooms + 𝛽2 * Bathrooms + 𝛽3 * Accommodates +  
# 𝛽4 * Price + 𝛽5 * Review.Scores.Rating + 𝛽6 * neighb_id
model_dfmadrid <-lm(formula=Square.Meters~ Bedrooms+Bathrooms+Accommodates+Price+Review.Scores.Rating+neighb_id , 
                      data=df_madrid.train) 
 
paste("Adjusted R-squared:  0.7676  ")
summary(model_dfmadrid)
AIC(model_dfmadrid)
```

#### Función que busca los n aptos más proximos en un df con respecto al apto que indicamos como parámetro

```r
# Función que devuelve los n pisos más proximos en un dataframe df con respecto al que pasamos por parámetro
# df: dataframe
# apto: apto sobre el cual buscar los aptos más proximos
# n: nº de aptos más proximos a devolver
get_aptos_proximos <- function (df, apto, n){
    
    # Número de los principales componentes PCA, nos quedamos con los 6 primeros
    number_of_pca_components<-6
    
    # Calcular el PCA del df
    prdf <-prcomp(df,center = TRUE, scale. = TRUE)
    
    # Transformado del piso dado en dominio PCA
    t_apto <- predict(prdf, newdata = apto)
    # Piso dado con coordenadas PCA a matriz
    t_apto<-matrix(t_apto[1:number_of_pca_components],nrow=1)
    
    # Matriz transformada PCA
    Apc<-prdf$x[,1:number_of_pca_components]
    dist<-rep(NA, nrow(Apc))
    # Calculamos la distancia euclidea entre el piso dado en PCA con la matriz transformada PAC
    for (i in 1:nrow(Apc)){
        dist[i]<-sum((t_apto-Apc[i,])^2)
    }
    
    # Ordenamos las distancias 
    ordend <- sort(dist)
    ordend
    
    # Nos quedamos con los n primeros apartamentos eliminando el primero que es el dado
    # Puesto que serán los más próximos al apartamento dado
    # Calculamos la distancia de n = nº de apartamentos a devolver + 1 para despues eliminar el propio apto
    dn <- ordend[n+1]
    dn
    
    # Data frame con los n apartamentos más proximos
    df_naptos_result <- head(df[which(dist <= dn), ])
    

    # Eliminamos el primer registro que corresponde al propio apto
    df_naptos_result<- df_naptos_result[2:nrow(df_naptos_result), ]
    print(paste("Los", n, "apartamentos más próximos al apto dado son: "))
    df_naptos_result
    
    
}
```

```r
# Apartamento dado por parámetro
i_apto <-251   # indice del apto sobre el cual realizaremos la búsqueda
print(paste("El apartamento dado como parámetro: "))
apto <-df_madrid_pca[i_apto,]
apto

get_aptos_proximos(df_madrid_pca, apto, 5)
```

#### Función que busca los n aptos más proximos en un df con respecto al apto cuyo indice pasamos como parámetro

```r
# Función que busca los n aptos más proximos en un df con respecto al apto cuyo indice pasamos como parámetro
# df: dataframe
# i_apto: indice para obtener el apto sobre el cual buscar los aptos más proximos
# n: nº de aptos más proximos a devolver
get_proximosaptos <- function (df, i_apto, n){
    
    # Número de los principales componentes PCA, nos quedamos con los 6 primeros
    number_of_pca_components<-6
    
    # Calcular el PCA del df
    prdf <-prcomp(df,center = TRUE, scale. = TRUE)
    
    # Obtenemos el apto base sobre el que comparar
    apto <- df[i_apto, ]
    
    # Transformado del piso dado en dominio PCA
    t_apto <- predict(prdf, newdata = apto)  
    # Piso dado con coordenadas PCA a matriz 
    t_apto<-matrix(t_apto[1:number_of_pca_components],nrow=1)
    
    # Matriz transformada PCA
    Apc<-prdf$x[,1:number_of_pca_components]
    dist<-rep(NA, nrow(Apc))
    # Calculamos la distancia euclidea entre el piso dado en PCA con la matriz transformada PAC
    for (i in 1:nrow(Apc)){
        dist[i]<-sum((t_apto-Apc[i,])^2)
    }
    
    # Ordenamos las distancias 
    ordend <- sort(dist)
    ordend
    
    # Nos quedamos con los n primeros apartamentos eliminando el primero que es el dado
    # Puesto que serán los más próximos al apartamento dado
    # Calculamos la distancia de n = nº de apartamentos a devolver + 1 para despues eliminar el propio apto
    dn <- ordend[n+1]
    dn
    
    # Data frame con los n apartamentos más proximos
    df_naptos_result <- head(df[which(dist <= dn), ])
    

    # Eliminamos el primer registro que corresponde al propio apto
    df_naptos_result<- df_naptos_result[2:nrow(df_naptos_result), ]
    print(paste("Los", n, "apartamentos más próximos al apto dado son: "))
    df_naptos_result
    
}
```

```r
# Calcula los 5 apartamentos más próximos respecto a apto = df_madrid_pca[8, ]
apto<-df_madrid_pca[8, ]
print(paste("El apartamento dado como parámetro: "))
apto
get_proximosaptos(df_madrid_pca, 8, 5)
```
